```python
import pandas as pd
import os

print(os.getcwd())
print(os.listdir())

df = pd.read_csv("flyrank-ml-internship/data/raw/content_refresh_anonymized.csv")

df["is_declining_label"] = (
    df["trend_direction"].str.lower() == "down"
).astype(int)

print("Total pages:", len(df))
print("Declining pages:", df["is_declining_label"].sum())
print("Declining rate:", round(df["is_declining_label"].mean() * 100, 2), "%")
print("Average CTR:", round(df["ctr"].mean(), 4))
```

```python
correlation = df['search_volume'].corr(df['ctr'])
print(f"Correlation between search_volume and CTR: {correlation:.4f}")
```
